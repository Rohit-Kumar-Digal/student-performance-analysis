```python
df["Total"] = df["Math"] + df["English"] + df["Science"]
# print(df["Total"])
```

```python
df["Average"] = df["Total"] / 3
# print(df[["Name","Average"]])?
```

```python
df.sort_values("Total",ascending=False).head()
```

```python
df.groupby("Department")["StudentID"].count()
```

```python
df.groupby("Department")[["Math","Science","English"]].mean()
```

```python
df[df["Attendance"]>90].head()
```

```python
df["Percentage"] = (df["Total"] / 300)*100
print(df["Percentage"].head())
```

```python
df["Result"] = np.where(df["Percentage"]>= 60,"Pass","Fail")
print(df[["Name","Result"]].head())
```

```python
df["Grade"] = np.where(df["Percentage"] >= 95,"A+",
              np.where(df["Percentage"] >= 90,"A",
              np.where(df["Percentage"] >= 80,"B",
              np.where(df["Percentage"] >= 60,"C","F"))))
print(df[["Name","Total","Percentage","Grade"]].head())
```

```python
import matplotlib.pyplot as plt
df.groupby("Department")["Total"].mean().plot(kind="bar")
plt.title("Average Mark By Department")
plt.xlabel("Department")
plt.ylabel("Average Mark")

plt.show()
```
